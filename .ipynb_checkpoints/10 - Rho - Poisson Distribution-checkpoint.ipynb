# HyperGeometric Probability Model


## What does X count?
The number of events or occurences within a specified time or space.

You need to keep track of units, i.e. the number of occurences per unit of time (or space).

## Application
Estimating the number of arrivals, or events that occur, in a fixed period of time or a fixed amount of space.

People in the know say, "X has a Poisson distribution with mean $\lambda$.

### Approximating a Poisson with a Normal Distribution Model

For large valuese of $\lambda$ (greater than 20) you can substitute the normal distribution for the Poisson.




## What's Fixed?


## Conditions
1. X counts the number of events or occurences within a specified time or space
2. The events occur independent of each other
3. No two events can happen at exactly the same time
4. Poisson must start at 0


## PMF

$\huge P(x) = \frac{e^{-\lambda}\lambda^x}{x!}$

$\large \lambda$: the average (mean) rate of occurance of the event over the fixed time or space

## CDF

$\Large F(x) = \displaystyle\sum_{X \geq x} \frac{e^{-\lambda}\lambda^x}{x!}$


## Expected Value

$\Large E(X) = \mu_x = \lambda$



## Variance & Standard Deviation

$\Large \sigma^2 = \lambda$

$\Large \sigma = \sqrt\lambda$
