## Carnival Game

A carnival game allows you three attempts to hit a target. You can hit or miss on each attempt. To win, you need to 'hit' twice in a row.

You are required to throw with alternate hands so if you start with your dominant hand, you have to throw with your weak hand on the second turn before returning to the dominant hand on the last throw.

You have a .70 probability of hiting with your dominant hand.
You have a .40 probability of hiting with your weak hand.

What is your probability of wining?



```python
# Answer:
# Two ways to win: HH, MHH

HH_prob = round(.7 * .4,3)
MHH_prob = round(.3*.4*.7,3)

Total_prob = round(HH_prob + MHH_prob,3)

Total_prob
```

2. Given that the current flowed from A to B, what is the probability that relay 1 functioned?

### 1 Answer:

This is a union problem. It's actually two unions. We are looking first at the P of '1 or 2'. Then we are also looking at the P of '3 or (1 or 2)'. lets look at '1 or 2' first.

$P(1 \cup 2) = P(1) + P(2) - P(1 \cap 2)$

$P(1 \cap 2)$

```python
one_and_two_prob = .9 * .9

one_and_two_prob
```

```python
one_or_two_prob = .9 + .9 - one_and_two_prob
one_or_two_prob
```

```python
three_and_one_or_two_prob = .9 * one_or_two_prob
three_and_one_or_two_prob
```

```python
three_or_one_or_two_prob = round(.9 + one_or_two_prob - three_and_one_or_two_prob,3)
three_or_one_or_two_prob
```
